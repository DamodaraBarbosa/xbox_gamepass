```python
import pandas as pd
import numpy as np
import re
```

```python
def clean_names(game: str):
    name = re.sub(r'[^a-zA-Z0-9\s]', '', game)
    return name.upper()
```

```python
master_list = pd.read_csv(r'data/mastelist.csv')
master_list['Game'] = master_list['Game'].apply(clean_names)
master_list.head(3)
```

```python
wiki_xbox_series = pd.read_csv(r'data\xbox_series_wikipedia.csv')
wiki_xbox_series['name'] = wiki_xbox_series['name'].apply(clean_names)
wiki_xbox_series.head(3)
```

```python
wiki_xbox_one = pd.read_csv(r'data\xbox_one_wikipedia.csv')
wiki_xbox_one['name'] = wiki_xbox_one['name'].apply(clean_names)
wiki_xbox_one.head(3)
```

```python
wiki_xbox_360 = pd.read_csv(r'data\xbox_360_wikipedia.csv')
wiki_xbox_360['name'] = wiki_xbox_360['name'].apply(clean_names)
wiki_xbox_360.head(3)
```

```python
developers_list = pd.read_csv(r'data\developers_list.csv')
developers_list.head()
```

# 2. ***Merge*** dos dados

```python
wiki_xbox = pd.concat([wiki_xbox_series, wiki_xbox_one, wiki_xbox_360])
```

```python
wiki_xbox = wiki_xbox.drop_duplicates(subset='name', keep='first')
wiki_xbox.index = range(wiki_xbox.shape[0])
```

```python
gamepass = pd.merge(
    left=master_list, right=wiki_xbox, how='left',
    left_on='Game', right_on='name'
).drop(columns='name')
```
